```python
#variabili = contenitori dinamici di dati aventi un loro tipo specifico

#collezione = collections = struttura dati che ci consente di gestire, organizzare e manipolare gruppi di dati
#liste, tuple, set, dizionari

#lista = struttura dati ordinata e modificabile
#possiamo accedere ad un elemento della lista tramite un indice e possiamo modificarla
#modificabile = acggiungere, rimuoovere e/o concatenare gli elementi

numeri = [10, 20, 30, 40, 50] #la lista si definisce con le parentesi quadre
print(numeri)

#indexing
#ogni elemento della lista ha un indice posizionale che parte da 0
print(numeri[0])
print(numeri[4])
#indice negativo = parto da destra a sinistra
print(numeri[-2])
#gli indici positivi partono da 0, mentre i negativi da -1

#le liste sono mutabili, quindi tramite indice posso modificare un valore presente
numeri[1] = 45
print(numeri)

#slicing
#prende i valori di un dato intervallo
#nomeLista[start:stop]
#indice di start è compreso, lo stop no
print(numeri[1:4]) #prenderà elementi con indice 1,2 e 3, il 4 è escluso
print(numeri[2:]) #con questa sintassi si prende da 2 fino alla fine
print(numeri[:3]) #prendo dall'inizio fino al 3 escluso

#step
#l'indice parte da 0 fino a 5 ecluso (tutti), ma con step 2, passo 2, quindi uno si e uno no
print(numeri[0:5:2]) # 0 2 4
#slicing: indice di fine è escluso, lo slicing affetta la collezione, taglia tra gli elementi

#le liste sono mutabili
numeri[4] = 100 #sto modificando un valore nella lista accedendo con indice
numeri.append(500) #con append sto aggiungendo in coda

print(numeri)

#stringhe
titolo = 'Il signore degli Anelli'
print(titolo [0]) #restituisce il primo carattere
print(titolo[-2])
#slicing
print(titolo[3:10]) #da indice tre a indice 10 escluso. Lo spazio conta come carattere

print(titolo.upper()) #restituisce tutto maiuscolo
print(titolo.lower()) #restituisce tutto minuscolo

nuovoTitolo = titolo.replace('degli Anelli', 'dei fornelli') #permette di sostituire una parte di stringa
print(nuovoTitolo)

#split mi consente di suddividere una stringa sulla base di un delimitatore
parole = titolo.split(' ') #suddivido il titolo in più parole ogni volta che incontra uno spazio
print(type(parole))

#conto quanti elementi della collezione
quanteParole = len(parole)
print(quanteParole)
```
