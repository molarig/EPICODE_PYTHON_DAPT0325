```python
# un costrutto logico condizionale è un'istruzione che permette di eseguire blocchi di codice a seconda delle condizioni.
#if condizione: --> è l'intestazione
#esegui azione (o più azioni) ---> è il corpo (con indentazione, ossia il tab per spostare il corpo più in la)
```

```python
x = 5
y = 7
if x < y:
    print('x è minore di y')
```

```python
x = 15
y = 7
if x < y:
    print('x è minore di y')
else:
    print('x è maggiore di y')
```

```python
cybersec = 10
data_analyst = 20
frontend = 30
backend = 40
```

```python
nuova_iscrizione = 'cybersec'

if nuova_iscrizione == 'cybersec':
    cybersec += 1
elif nuova_iscrizione == 'data_analyst':
    data_analyst += 1
elif nuova_iscrizione == 'frontend':
    frontend += 1
elif nuova_iscrizione == 'backend':
    backend += 1

print('cybersec', cybersec)
print('data_analyst', data_analyst)
print('frontend', frontend)
print('backend', backend)
```

```python
#i loop servono a eseguire dei blocchi di codice in successione
#con while, il costrutto esegue sempre le stesse istruzioni finchè è vera una determinata condizione
```

```python
#sommo i numeri da 1 a 15
num = 1 #inizializzazione
somma = 0 #inizializzazione
while num <= 15:
    somma += num # calcolo
    print('num', num, '-somma =', somma)
    num += 1 #condizione di aggiornamento
print('la somma è', somma)
#le istruzioni indentate possono essere eseguite 0, 1 o più volte, quelle non indentate vengono eseguite sempre
```

```python
somma = 0

while 0 <= somma <= 50:
    num = input('Inserisci un numero: ')
    num = int(num)
    somma += num

print('La somma è:', somma)
```

```python
#for significa per, nel senso di 'per ognuno' o 'per tot di volte'
prezzi = [10, 20, 30, 40, 50]

for prezzo in prezzi:
    print ('-', prezzo * 0.8)
```
